```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

# Preprocess and Split Dataset

* In regression models, we use the mean value - **avg**, to predict the feasibility of the classification model.
* the accuracy that you see in regression models is the highest accuracy we got
* And the **avg** in classification models is defined below 


```python
# Xtr = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/dataset.csv")
Xtr = pd.read_csv("/content/dataset.csv")
h = Xtr.columns.to_list()
rh = ['popularity'] + [str(it) for it in range(100)]
h = [it for it in h if it not in rh]
print(h)
Xtr.drop(h, axis=1, inplace=True)

Al = Xtr['popularity']
Al = np.array(Al)
Al.sort()
idx = len(Al)*0.25
print(idx)
print(Al[-50])
avg = Al[-50]

X_train, X_test, y_train, y_test = train_test_split(Xtr, Xtr['popularity'], test_size=0.35, random_state=42)
X_train.drop(['popularity'] , axis=1, inplace=True)
X_test.drop(['popularity'] , axis=1, inplace=True)
X_train, X_test, y_train, y_test = np.array(X_train), np.array(X_test), np.array(y_train), np.array(y_test)
print( X_train.shape, X_test.shape, y_train.shape, y_test.shape )

# for classification training
print(avg)
y1, y2 = [], []
for it in y_train:
  if it > avg:
    y1.append(1)
  else:
    y1.append(0)

for it in y_test:
  if it > avg:
    y2.append(1)
  else:
    y2.append(0)
y_train = np.array(y1)
y_test = np.array(y2)
```

# Sequential Neural Networks Training - Regressor

```python
from tensorflow import keras
from tensorflow.keras import layers
from keras.layers import Dense, Input, LeakyReLU

model = keras.Sequential([
    layers.Dense(64, activation=LeakyReLU(alpha=0.01), kernel_initializer = "he_normal", input_shape=(X_train.shape[1],)),
    layers.Dense(32, activation=LeakyReLU(alpha=0.01),  kernel_initializer = "he_normal"),
    layers.Dense(32, activation=LeakyReLU(alpha=0.01),  kernel_initializer = "he_normal"),
    layers.Dense(1, activation=LeakyReLU(alpha=0.01), kernel_initializer = "he_normal")
    ])
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=50, batch_size=32)
mse = model.evaluate(X_test, y_test, verbose=0)
print(f'MSE: {mse}')
```

```python
predictions = model.predict(X_test)
predictions = list(predictions)
predictions = [list(it) for it in predictions]
for i in range(len(predictions)):
  predictions[i].append(y_test[i])
print(predictions)
s = 0
for it in predictions:
  p1 = 1 if it[0] > avg else 0
  p2 = 1 if it[1] > avg else 0
  if p1 == p2 : s += 1
acc = s/len(predictions)
print("Binary Classification Acc：", acc)
```

# DTree Training - Regressor

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error

regressor = DecisionTreeRegressor()
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print(f"mse: {mse:.2f}")
cmp = [(y_test[i],y_pred[i]) for i in range(len(X_test))]
print(cmp)
s = 0
for it in cmp:
  p1 = 1 if it[0] > avg else 0
  p2 = 1 if it[1] > avg else 0
  if p1 == p2 : s += 1
acc = s/len(cmp)
print("Binary Classification Acc：", acc)
```

# Random Forest - Regressor

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

regressor = RandomForestRegressor()
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print(f"mse: {mse:.2f}")
cmp = [(y_test[i],y_pred[i]) for i in range(len(X_test))]
print(cmp)
s = 0
for it in cmp:
  p1 = 1 if it[0] > avg else 0
  p2 = 1 if it[1] > avg else 0
  if p1 == p2 : s += 1
acc = s/len(cmp)
print("Binary Classification Acc：", acc)
```

# Sequential Neural Networks Training - Binary Classification

```python
from tensorflow import keras
from tensorflow.keras import layers
from keras.layers import Dense, Input, LeakyReLU

model = keras.Sequential([
    layers.Dense(64, activation=LeakyReLU(alpha=0.01), kernel_initializer = "he_normal", input_shape=(X_train.shape[1],)),
    layers.Dense(32, activation=LeakyReLU(alpha=0.01),  kernel_initializer = "he_normal"),
    layers.Dense(32, activation=LeakyReLU(alpha=0.01),  kernel_initializer = "he_normal"),
    layers.Dense(1, activation='sigmoid', kernel_initializer = "he_normal")
    ])
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=50, batch_size=32)
mse = model.evaluate(X_test, y_test, verbose=0)
print(f'Loss and Accuracy: {mse}')
```
