```python
from audiodiffusion.audio_encoder import AudioEncoder

audio_encoder = AudioEncoder.from_pretrained("teticio/audio-encoder")
```

Load dataset and encode

```python
import pandas as pd
import numpy as np
import os
path = '/content/drive/MyDrive/test'
outputpath = '/content/drive/MyDrive/Colab Notebooks/test_dataset.csv'

df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/test_music.csv")
h = df.columns.to_list()
h = h + (list(range(100)))
data = np.array(df)

train_dataset = []
for it in data:
  name = it[1]
  name = (name+'.mp3').replace('?','_')
  name = name.replace(':','_')
  name = name.replace('|','_')
  name = name.replace('"','_')
  name = name.replace('/','_')
  directory = os.path.join(path,name)
  if not os.path.exists(directory):
    print(name,'不存在')
  else:
    item0 = list(it) + audio_encoder.encode([directory]).numpy()[0].tolist()
    train_dataset.append(item0)
    print(item0)
df = pd.DataFrame(train_dataset)
df.to_csv(outputpath, sep=',', index=False, header=False)
df = pd.DataFrame(train_dataset, columns=h)
df.to_csv(outputpath,sep=',',index=False,header=True)
```
