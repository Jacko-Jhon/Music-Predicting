```python
import os
import pandas as pd
import numpy as np

musics = []
for i in range(200):
    musics.append(music_name[i])
mu = []

s = 0
path = r'E:\music'
for name in musics:
    name = (name+'.mp3').replace('?','_')
    name = name.replace(':','_')
    name = name.replace('|','_')
    name = name.replace('"','_')
    name = name.replace('/','_')
    directory = os.path.join(path,name)
    if not os.path.exists(directory):
        print(name,'not exist')
        s += 1


print("unfind music: ",s)
print("total music: ",len(musics))
```

## Randomly select songs

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df1 = pd.read_csv('./music.csv')
ids = np.array(df1['id'])
# data = [df.loc[i] for i in range(53,100)]
df = pd.read_csv('./archive/tracks.csv')
df = df[df['release_date'] > '1999']
head = df.columns.to_list()
print(head)
```

```python
import random
m = len(df)
print(m)
dfnp = np.array(df)
print(dfnp[0])
ds = [it for it in dfnp if it[0] not in ids]
m = len(ds)
print(m)

data = []
for i in range(m):
    rdint = random.randint(1,222296)
    if rdint <= 200:
        data.append(ds[i])

print(len(data))
print(data)
```

```python
head = np.array(head)
csvdata = pd.DataFrame(data, columns=head)
csvdata.to_csv('music1.csv', index=False, header=True)
```

```python
from datetime import datetime
import numpy as np
data = np.array(data)
# popularity = data[:,2]
popularity = data[:,2]
# release_year = data[:,7]
release_year = data[:,7]
release_year = [str(it) for it in release_year]

datetimes = []
for date_str in release_year:
    dt = None
    if len(date_str) == 4:  # Year-only format (e.g., '1925')
        dt = datetime.strptime(date_str, '%Y')  # Parse year
    elif len(date_str) == 7:  # Year-month format (e.g., '1920-10')
        dt = datetime.strptime(date_str, '%Y-%m')  # Parse year and month
    elif len(date_str) == 10:  # Year-month-day format (e.g., '1920-01-02')
        dt = datetime.strptime(date_str, '%Y-%m-%d')  # Parse year, month, and day
    datetimes.append(dt)

plt.figure(figsize=(20,10))
plt.scatter(datetimes, popularity)
plt.xlabel('Release Year')
plt.ylabel('Popularity')
plt.show()
```
